#### Regner ut idle time for fuel 1-5 (ikke strøm)

```python
import csv

folder = 'generated_datafiles_allroutes/'
sluttfilnavn = ''

# Function to calculate duration_idle based on given rules
def calculate_duration_idle(sailing, lossing):
    result = []
    for sail, loss in zip(sailing, lossing):
        sail = float(sail[0])  # Extracting the value from the list
        loss = float(loss[0])  # Extracting the value from the list
        total = sail + loss

        if total < 24:
            result.append(24 - total)
        elif total < 48:
            result.append(48 - total)
        elif total < 72:
            result.append(72 - total)
        else:
            result.append(96 - total)
    return result

# Read duration_sailing.csv
with open(f'{folder}duration_sailing{sluttfilnavn}.csv', 'r') as sailing_file:
    sailing_reader = csv.reader(sailing_file)
    sailing_data = list(sailing_reader)[1:]  # Skip the header row

# Read duration_lossing.csv
with open(f'{folder}duration_lossing{sluttfilnavn}.csv', 'r') as lossing_file:
    lossing_reader = csv.reader(lossing_file)
    lossing_data = list(lossing_reader)[1:]  # Skip the header row

# Calculate duration_idle
duration_idle = calculate_duration_idle(sailing_data, lossing_data)

# Write duration_idle to a new CSV file
with open(f'{folder}duration_idle{sluttfilnavn}.csv', 'w', newline='') as idle_file:
    writer = csv.writer(idle_file)
    writer.writerow(['duration_idle'])
    writer.writerows(map(lambda x: [x], duration_idle))
```

#### Regner ut antall megajoules per rute for fuel 1-5 (ikke strøm)

```python
import csv

# Function to calculate mj_route based on given durations
def calculate_mj_route(sailing, lossing, idle):
    result = []
    for sail, loss, idle_duration in zip(sailing, lossing, idle):
        sail = float(sail[0])  # Extracting the value from the list
        loss = float(loss[0])  # Extracting the value from the list
        idle_duration = float(idle_duration[0])  # Extracting the value from the list

        mj_value = sail * 9000 + loss * 2160 + idle_duration * 1000
        result.append(round(mj_value, 2))  # Round to two decimal places
    return result

# Read duration_sailing.csv
with open(f'{folder}duration_sailing{sluttfilnavn}.csv', 'r') as sailing_file:
    sailing_reader = csv.reader(sailing_file)
    sailing_data = list(sailing_reader)[1:]  # Skip the header row

# Read duration_lossing.csv
with open(f'{folder}duration_lossing{sluttfilnavn}.csv', 'r') as lossing_file:
    lossing_reader = csv.reader(lossing_file)
    lossing_data = list(lossing_reader)[1:]  # Skip the header row

# Read duration_idle.csv
with open(f'{folder}duration_idle{sluttfilnavn}.csv', 'r') as idle_file:
    idle_reader = csv.reader(idle_file)
    idle_data = list(idle_reader)[1:]  # Skip the header row

# Calculate mj_route
mj_route = calculate_mj_route(sailing_data, lossing_data, idle_data)

# Write mj_route to a new CSV file
with open(f'{folder}mj_route{sluttfilnavn}.csv', 'w', newline='') as mj_route_file:
    writer = csv.writer(mj_route_file)
    writer.writerow(['mj_route'])
    writer.writerows(map(lambda x: [x], mj_route))
```
